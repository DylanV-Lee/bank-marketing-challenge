The variables are not resembling what they are meant to.

```python
# create a list of feature that are integer

integer_columns = ['Age', 'Balance', 'Duration', 'Campaign', 'Pdays', 'Previous']

# converting the columns to interger
new_data[integer_columns] = new_data[integer_columns].apply(pd.to_numeric, errors = 'coerce').astype('Int64')
```

```python
new_data.dtypes
```

```python
# identifying the columns with object datatype

object_columns = new_data.select_dtypes(include = 'object').columns

# capitalize the first letter of the words

new_data[object_columns] = new_data[object_columns].apply(lambda x: x.str.capitalize())
```
